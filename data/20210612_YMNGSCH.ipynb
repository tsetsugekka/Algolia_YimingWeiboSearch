```python
import pdfplumber
import pandas as pd
import requests,re,time,os
import json,ast
import numpy as np
```

```python
path="C://Users//tong//Desktop//张一鸣微博日记2286条2019-05-25.pdf"
vialist=["今日头条","iPhone6Plus","手机微博触屏版","iPhone客户端","微博 weibo.com"]
jsondata=[]

with pdfplumber.open(path) as pdf: 
    
    maxpage=len(pdf.pages)
    content=""

    for i in range(1,maxpage):
        page = pdf.pages[i] 
        page_content = '\n'.join(page.extract_text().split('\n')[:-1])
        page_content=page_content.replace("@张一鸣","@一鸣")
        content=content+page_content

    weibos=content.split('张一鸣\n')[1:]
    
    for weibo in weibos:
        weibo.replace("@一鸣","@张一鸣")
        status=weibo.split('\n')[0].split('来自')
        time=status[0][-6:]
        year=status[0][:4]
        date=status[0][:-6]    
        tweet="".join(weibo.split('\n')[1:])
        
        if len(status[1])<20:
            via=status[1]
        else:
            for i in vialist:
                if i in status[1]:
                    via=i
                    tweet=status[1].replace(i,"")+tweet
        
        jsondata.append([date,year,time,via,tweet])
        
df=pd.DataFrame(jsondata,
            columns=['Date', 'Year','Time','Via','WeiboText'])
df
```

```python
out=df.to_json(orient='records',force_ascii=False)
with open("20210612_Poject_YMNGSCH.json", 'w', encoding='utf-8') as outfile:
    outfile.write(out)
```
